# Model Architecture

```python
from tensorflow.keras import layers
from keras.models import Sequential
from keras.layers import Dense, GlobalAveragePooling2D, Input
import tensorflow as tf
from tensorflow.keras.applications.xception import Xception

IMG_HEIGHT = 150  # Decrease if needed
IMG_WIDTH = 150   # Decrease if needed
IMG_CHANNELS = 3

input_shape = (IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS)

def xception_transfer_actual(input_shape):
    res = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    for layer in res.layers:
        layer.trainable = True

    model = Sequential()
    model.add(res)
    model.add(GlobalAveragePooling2D())  # Replace Flatten with Global Average Pooling
    model.add(Dense(units=999, activation="relu"))
    model.add(Dense(units=499, activation="relu"))  # Reduced number of units
    model.add(Dense(units=99, activation="relu"))
    model.add(Dense(units=4, activation="softmax"))  # Adjust output units to match your classes

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    model.summary()

    return model

ptXceptionModel = xception_transfer_actual(input_shape)
```

```python
from sklearn.metrics import roc_curve, auc
import seaborn as sns

# Visualise the ROC curve model performance for deep learning
def plot_multiclass_roc_dcnn(clf, xTest, yTest, n_classes, figsize=(17, 6)):
    y_score = clf.predict(xTest)
    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(np.argmax(yTest, axis=1), drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic (ROC)')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

```python
import seaborn as sns

# Overall model performance
def model_performance(y_val, y_pred_prob, model_name, classes):
    # Confusion matrix and classification report
    cm = confusion_matrix(y_val, y_pred_prob.argmax(axis=1))
    print("--- Confusion Matrix ---")
    print(cm)

    # Plot the confusion matrix figure
    df_cm = pd.DataFrame(cm, range(len(classes)), range(len(classes))) # Use the length of 'classes' to define shape
    sns.set(font_scale=1.4)
    sns.heatmap(df_cm, annot=True, annot_kws={"size": 16})
    plt.show()

    print("--- Classification Report ---")
    print(classification_report(y_val, y_pred_prob.argmax(axis =1), target_names=classes))

    # Plot F1-score vs. Classes

    f1Score = f1_score(y_val, y_pred_prob.argmax(axis=1), average=None)
    y_pos = np.arange(len(classes))
    plt.bar(y_pos, f1Score)
    plt.xticks(y_pos, classes)
    plt.ylabel('F1 Score')
    plt.title('Final F1 Score of various species after classification')
    plt.show()
    return cm, f1Score
```
